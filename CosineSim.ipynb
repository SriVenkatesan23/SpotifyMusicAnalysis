```python
import collections, re
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from scipy import spatial
```

```python
vectorizer = CountVectorizer(stop_words = 'english')
```

```python
songDf = pd.read_csv("songData.csv")
songDf
```

```python
emptyList = (pd.isnull(songDf['lyric']))
notEmptyList = list(~np.array(emptyList))
notEmptyIndex = np.where(notEmptyList)[0]
songDf['lyric'][notEmptyIndex];
```

```python
name = songDf['name'][notEmptyIndex];
artist = songDf['artist'][notEmptyIndex];
```

```python
data_corpus = songDf['lyric'][notEmptyIndex]
```

```python
X = vectorizer.fit_transform(data_corpus)
```

```python
tdMatrix = pd.DataFrame(X.toarray())
tdMatrix.columns = vectorizer.get_feature_names()
```

```python
tfidfTransform = TfidfTransformer()
```

```python
tfidfValues = tfidfTransform.fit_transform(X)
tfidfMatrix = pd.DataFrame(tfidfValues.toarray())
tfidfMatrix.columns = vectorizer.get_feature_names()
```

```python
1-spatial.distance.cosine(np.array(tfidfMatrix.iloc[0]) , np.array(tfidfMatrix.iloc[38]))
```

```python
def cosSim(index):
    lyricSim = {}
    for x in range(len(tfidfMatrix)):
        sim = 1 - spatial.distance.cosine(np.array(tfidfMatrix.iloc[index]), np.array(tfidfMatrix.iloc[x]))
        lyricSim.update({x:sim})
    values = sorted(lyricSim.items(), key=lambda x:x[1])
    return(values)
```

```python
logicSim = cosSim(0);
```

```python
print(name.iloc[0] + ": " + artist.iloc[0])
print("")
for x in logicSim[8113:8120]:
    print(name.iloc[x[0]] + ": " + artist.iloc[x[0]] + " score: " + str(x[1]))
```
