```python
import pandas as pd

file = pd.read_csv("songdata.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = file
data['duration_ms'] = data['duration_ms'] / 1000
columns = list(file.select_dtypes(include=['number']).columns.values)

fig, axes = plt.subplots(1, 16, figsize=(100,10))

for i, col in enumerate(columns):
    if i < 2: continue
    data[col].hist(bins=np.linspace(pd.Series(data[col]).min(), pd.Series(data[col]).max()   ,10),ax=axes[i])
    axes[i].set_title('Histogram For {0}'.format(col.capitalize()))
    axes[i].set_xlabel(col.capitalize())
    if i == 4:
        axes[i].set_title('Histogram For Duration')
        axes[i].set_xlabel("Seconds")
    axes[i].set_ylabel('Count')

fig.tight_layout()
fig.savefig('test.jpg')
```

```python
data = file
columns = list(file.select_dtypes(include=['number']).columns.values)
columns = columns[2:]

d = {'col1': [1, 2], 'col2': [3, 4]}

dicts = {}
keys = ['Feature', 'Mean', 'Median', 'Variance', 'Range']

for i in keys:
    if(i=='Feature'):
        dicts[i] = columns
    else:
        ls = list()
        if(i=='Mean'):
            for s in columns:
                ls.append(file[s].mean())
            dicts[i] = ls
        elif(i=='Median'):
            for s in columns:
                ls.append(file[s].median())
            dicts[i] = ls
        elif(i=='Variance'):
            for s in columns:
                ls.append(file[s].var())
            dicts[i] = ls
        elif(i=='Range'):
            for s in columns:
                ls.append(file[s].max() - file[s].min())
            dicts[i] = ls


pd.DataFrame(dicts)





```
